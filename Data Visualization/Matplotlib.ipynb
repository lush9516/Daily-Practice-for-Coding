# Bar plot

### Basic bar

```python
plt.figure(figsize=(4,4))
# determine the width & height of the figure

plt.bar([1,2,3], [2,3,7],align='center', alpha=0.5)
# first list determine the position for each in the x-axis
# second list determine the height of each bar
# yerr or xerr are used to generate error bars
plt.grid()
# show the grid in the plot

# set axes labels and title
plt.ylabel('variable y')
plt.xticks([1,2,3],['bar1','bar2','bar3'])
# first list is the position of the labels
# second determine the labels
plt.title('Bar plot with error bars')

plt.ylim([0,10])
plt.xlim([0,4])
# xlim & ylim are used to set range for x-axis &y-axis
plt.show()
#plt.savefig('./my_plot.png')
```

### Grouped bar 

#### Basic idea: Setting the x-position argument by using ([i + width for i in x-pos],width) for 2nd bar
#### and ([i + width*2 for i in x-pos],width) for the 3rd bar and etc.

```python
x_pos = [1,2,3]
width =0.3
plt.bar(x_pos,[1,2,5],width,alpha=0.5,color='r',label=['group1'])
plt.bar([i + width for i in x_pos],[2,4,10], width,alpha=0.5,color = 'b',label=['group2'])
plt.bar([i + width*2 for i in x_pos],[3,6,15],width,alpha=0.5,color = 'y',label=['group3'])
plt.legend(['lowest', 'middle', 'highest'], loc='upper left')
plt.xticks([1.25,2.25,3.25],['bar1','bar2','bar3'])
# first arg of xticks is the location of x-labels and the 2nd arg of xticks is the name of x-labels

# 注释标签_ plt.legend_set label for each categoty using label = 
plt.grid()
```

# Histogram plot

### Basic histogram

```python
data = np.random.normal(0, 20, 10000) 

plt.hist(data, bins=50, alpha=0.5)
plt.title('Random Gaussian data (fixed number of bins)')
plt.xlabel('variable X (20 evenly spaced bins)')
plt.ylabel('count')
plt.xlim([min(data)-5, max(data)+5])

plt.show()
```

### Histogram with two overlapping data

#### Just plot the two histograms 

```python

data1 = np.random.normal(0, 20, 10000) 
data2 = np.random.normal(50, 20, 10000) 

plt.hist(data1, bins=50, alpha=0.3, label='class 1')
plt.hist(data2, bins=50, alpha=0.3, label='class 2')
plt.title('Random Gaussian data')
plt.xlabel('variable X')
plt.ylabel('count')
plt.legend(loc='upper right')
```

# Line plot

### Simple plot

```python
x = [1, 2, 3]

y_1 = [50, 60, 70]
y_2 = [20, 30, 40]

plt.plot(x, y_1, marker='x',color='y',alpha = 0.6,label='sample1')
plt.plot(x, y_2, marker='^',color='b',alpha = 0.4,label='sample2')

plt.xlim([0, len(x)+1])
plt.ylim([0, max(y_1+y_2) + 10])
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('Simple line plot')
plt.legend(loc='upper left')

plt.show()
```

### Line plot with error bars

```python
x=[1,2,3]
y1 = [2,4,10]
y2 = [3,12,20]
y1_err = [0.5,2.5,2]
y2_err = [0.2,2.7,4.6]

plt.errorbar(x, y1, yerr=y1_err, fmt='-x')
plt.errorbar(x, y2, yerr=y2_err, fmt='-^')
plt.xlim([min(x)-2, max(x)+2])
plt.ylim([min(y1)-2, max(y2)+10])
plt.show()
```

### Step plot

```python
A = [1,2,3,4,5,6,7]
B = [1,3,6,12,19,30,50] # 10 rand. values from a std. norm. distr.
C = np.array(B).cumsum()
# cumsum() can only be used for array
plt.step(A,C)

plt.ylabel('sample count')
plt.xlabel('x value')
plt.title('Number of samples at a certain threshold')

plt.show()
```
